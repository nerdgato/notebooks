# LIMPIAR VENTAS

Para el siguiente proceso se dejara el formato dd-mm-yyyy ya que es mas legible en español y se eliminara la hora:minutos:segundos ya que al ser igual en todas las filas (00:00:00) no nos sirve de mucho.

Ademas se normalizara la id de los clientes a simples numeros enteros, en vez de utilizar el prefijo "CLI" y el maximo de 5 digitos. (¿Que ocurrira cuando se excedan esos 5 digitos y pase a 6 digitos? Ademas probablemente haya que implementar un algoritmo que rescate el digito, lo haga autoincrementable y nuevamente incorporarle el CLI y los ceros)



```python
df_original['DOCUMENTO'] = df_original['DOCUMENTO'].str.extract('(\d+)').astype(int)


df_original['FECHA'] = pd.to_datetime(df_original['FECHA'])


df_original['FECHA'] = df_original['FECHA'].dt.strftime('%d-%m-%Y')


df_original['ID_CLIENTE'] = df_original['ID_CLIENTE'].str.extract('(\d+)').astype(int)


# Verificamos el resultado
df_original.head(10)
```

# Todo el proceso junto y exportacion

```python


ruta_carpeta = 'big_data'
nombre_archivo = 'Americans.xlsx'
ruta_archivo = os.path.join(ruta_carpeta, nombre_archivo)


ventas = pd.read_excel(ruta_archivo , sheet_name='VENTAS')

ventas['DOCUMENTO'] = ventas['DOCUMENTO'].str.extract('(\d+)').astype(int)
ventas['FECHA'] = pd.to_datetime(ventas['FECHA'])
ventas['FECHA'] = ventas['FECHA'].dt.strftime('%d-%m-%Y')
ventas['ID_CLIENTE'] = ventas['ID_CLIENTE'].str.extract('(\d+)').astype(int)


clientes = pd.read_excel(ruta_archivo , sheet_name='CLIENTES')

clientes['ID_CLIENTE'] = clientes['ID_CLIENTE'].str.extract('(\d+)').astype(int)


detalle_venta = pd.read_excel(ruta_archivo , sheet_name='DETALLE_VENTA')

detalle_venta['DOCUMENTO'] = detalle_venta['DOCUMENTO'].str.extract('(\d+)').astype(int)
detalle_venta['CODIGO_PRODUCTO'] = detalle_venta['CODIGO_PRODUCTO'].str.extract('(\d+)').astype(int)


sucursal = pd.read_excel(ruta_archivo , sheet_name='SUCURSAL')

sucursal['NOMBRE_SUCURSAL'] = sucursal['NOMBRE_SUCURSAL'].str.replace('Sucursal ', '', regex=False)

articulos = pd.read_excel(ruta_archivo , sheet_name='ARTICULOS')
articulos['CODIGO_PRODUCTO'] = articulos['CODIGO_PRODUCTO'].str.extract('(\d+)').astype(int)

canal = pd.read_excel(ruta_archivo , sheet_name='CANAL')
comuna = pd.read_excel(ruta_archivo , sheet_name='COMUNA')
provincia = pd.read_excel(ruta_archivo , sheet_name='PROVINCIA')
region = pd.read_excel(ruta_archivo , sheet_name='REGION')
vendedores = pd.read_excel(ruta_archivo , sheet_name='VENDEDORES')



nombre_archivo_salida = 'Americans_limpio.xlsx'
ruta_archivo_salida = os.path.join(ruta_carpeta, nombre_archivo_salida)

with pd.ExcelWriter(ruta_archivo_salida) as writer:
    articulos.to_excel(writer, sheet_name='ARTICULOS', index=False)
    canal.to_excel(writer, sheet_name='CANAL', index=False)
    clientes.to_excel(writer, sheet_name='CLIENTES', index=False)
    comuna.to_excel(writer, sheet_name='COMUNA', index=False)
    detalle_venta.to_excel(writer, sheet_name='DETALLE_VENTA', index=False)
    provincia.to_excel(writer, sheet_name='PROVINCIA', index=False)
    region.to_excel(writer, sheet_name='REGION', index=False)
    sucursal.to_excel(writer, sheet_name='SUCURSAL', index=False)
    vendedores.to_excel(writer, sheet_name='VENDEDORES', index=False)
    ventas.to_excel(writer, sheet_name='VENTAS', index=False)

print(f'Archivo exportado exitosamente en: {ruta_archivo_salida}')
```
